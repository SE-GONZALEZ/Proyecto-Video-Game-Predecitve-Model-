**Convertir los datos Strings a número** 

```python
#CONVERTIR STRINGS A NÚMEROS
label_encoder = preprocessing.LabelEncoder()
cols=["Platform","Genre","Publisher","Rating","Critic_Score_Class"]
for col in cols:
  dataset[col]=label_encoder.fit_transform(dataset[col])
dataset.head(10)


# Plataform: 0 = 3DS, 1 = DC, 2 = DS, 3 = GBA, 4 = GC, 5 = PC, 6 = PS, 7 = PS2, 8 = PS3, 9 = PS4, 
#            10 = PSP, 11 = PSV, 12 = Wii, 13 = WiiU, 14 = X, 15 = X360, 16 = XOne
# Genre: 0 = Action, 1 = Adventure, 2 = Fighting, 3 = Misc, 4 = Plaform, 5 = Puzzle, 6 = Racing, 7 = Role-Playing, 8 = Shooter,  
#        9 = PS4, 10 = Simulation, 11 = Sports, 12 = Strategy
# Rating: 0 = AO, 1 = E, 2 = E10+, 3 = K-A, 4 = M, 5 = RP, 6 = T
# Rating: 0 = Aceptable, 1 = Bueno, 2 = Excelente, 3 = Malo
```

**Tener en cuenta que tanto JP_Sales como NA_Sales al tener una ',' son tratados como objetos, entonces, toca remplazar dicha ',' por un punt '.' (Solamente esas 2 columnas ya que serán las únicas columnas sales que utilizaremos)**

```python
dataset['NA_Sales'] = dataset['NA_Sales'].str.replace(',','.').astype(np.float64)
dataset['JP_Sales'] = dataset['JP_Sales'].str.replace(',','.').astype(np.float64)
dataset['Platform'] = dataset['Platform'].astype(np.float64)
dataset['Genre'] = dataset['Genre'].astype(np.float64)
dataset['Rating'] = dataset['Rating'].astype(np.float64)
dataset['Critic_Score_Class'] = dataset['Critic_Score_Class'].astype(np.float64)
```

**Después de realizar la limpieza de datos, se vuelve a revisar el tipo de variable que es cada columna**

```python
dataset.info()
```

**Nuevo CSV posterior a la limpieza de datos**

```python
dataset = dataset.drop(["EU_Sales","Publisher","Other_Sales","Global_Sales"], axis=1)
```

```python
dataset
```

```python
dataset.to_csv("Archivos/DatasetFinal.csv",index=False)
```

**Determinación de variables Y y X**

```python
Y_1 = dataset.NA_Sales
#Y=dataset.NA_Sales
X = dataset.drop(["NA_Sales"], axis=1)
```

```python
lab = preprocessing.LabelEncoder()
Y = lab.fit_transform(Y_1)
```

```python
Y
```

```python
X.info()
```

**TRAINING Y TEST -> 80% ; VALIDATION -> 20%**

```python
train_size=0.8
test_size = 0.20
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.8)
```

**Escalización**

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

**PCA**

```python
PCA = decomposition.PCA(n_components=5)
PCA.fit(X_train)
X_train = PCA.transform(X_train)

print("Pesos n de PCA:",PCA.explained_variance_ratio_)
print("Suma total de pesos n de PCA",sum(PCA.explained_variance_ratio_))
```

# **SVM**

**Optimizar los hiperparametros utilizando Grid-search**

```python
param_grid=[{'kernel':["linear","poly","rbf"],'gamma':["scale","auto"],'C':[0.1,1,2.5,5,7.5,10,30,50,75,100]}]
gridsearchCV=GridSearchCV(estimator=SVR(),param_grid=param_grid, scoring="r2")
gridsearchCV.fit(X_train,y_train)
```

**Se muestran los resultados**

```python
gridsearchCV.cv_results_
```

**Los mejores parametros para obtener el mejor score**

```python
gridsearchCV.best_params_
```

**Resultado R2**

```python
gridsearchCV.best_score_
```
